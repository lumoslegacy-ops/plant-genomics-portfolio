```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV
df = pd.read_csv('CpG_Methylation_Data.csv')

# Show stats
print("Total CpG sites:", len(df))
print("Average methylation:", df['MethylationRatio'].mean())

# Plot
plt.figure(figsize=(8, 5))
plt.hist(df['MethylationRatio'], bins=50, color='mediumseagreen', edgecolor='black')
plt.title("CpG Methylation Distribution")
plt.xlabel("Methylation Ratio")
plt.ylabel("Number of CpG Sites")
plt.grid(True)
plt.tight_layout()
plt.show()
```
