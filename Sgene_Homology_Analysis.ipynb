```python
import matplotlib.pyplot as plt

# Real BLAST identity data (Arabidopsis vs Tomato)
genes = ["AtMLO2", "AtSWEET11", "AtDMR6", "AtPMR4"]
identities = [62.43, 53.60, 67.35, 74.75]

# Plotting
plt.figure(figsize=(6, 4))
bars = plt.bar(genes, identities)

# Add value labels on bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 1, f'{yval:.2f}%', ha='center', fontsize=9)

# Chart styling
plt.title("Identity % of Tomato Homologs to Arabidopsis S-Genes")
plt.xlabel("Arabidopsis S-gene")
plt.ylabel("Identity (%)")
plt.ylim(0, 80)
plt.tight_layout()

# Save high-resolution image for use in project
plt.savefig("Sgene_identity_real_blast.png", dpi=300)
plt.show()
```
